```python
import matplotlib.pyplot as plt
from tdmq.client import Client
```

```python
c = Client()
sources = c.find_sources(args={'entity_type': 'WeatherObserver'})
len(sources)
```

```python
s = sources[0]
s.description
```

```python
ts = s.timeseries(bucket=1000, op='avg')
dts = [_[0] for _ in ts]
temps = [_[1]['temperature'] for _ in ts]
plt.plot(dts, temps)
```
